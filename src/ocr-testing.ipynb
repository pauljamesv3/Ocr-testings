```python
import keras_ocr
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
images = [
    keras_ocr.tools.read(url) for url in [
        "/content/IMG_20240520_174246 (1).jpg",
        "/content/image_0.png",
        "/content/img_00.jpg",
        "/content/img_001.png"
    ]
]
```

```python
prediction_groups = pipeline.recognize(images)
prediction_groups
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```

```python
from lanyocr import LanyOcr
import time

ocr = LanyOcr(
        detector_name="paddleocr_en_ppocr_v3",
        recognizer_name= "paddleocr_en_ppocr_v3",
        merger_name= "lanyocr_nomerger",
        merge_boxes_inference= False,
        merge_rotated_boxes=True,
        merge_vertical_boxes=False,
        use_gpu=False,
        disable_angle_classifier=False,
        debug=False,
    )
```

```python
start = time.time()
results = ocr.infer_from_file("/content/IMG_20240520_174246 (1).jpg")
end = time.time()
print(end-start)
results
```
